```python
#Dependencies
import pandas as pd
import os
from bs4 import BeautifulSoup as bs
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

### NASA Mars News Scraping

```python
#Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#Read url
news_url = 'https://redplanetscience.com/'
browser.visit(news_url)

#Create BeautifulSoup object
news_html = browser.html
news_soup = bs(news_html,'html.parser')
```

```python
#Analyse the url page
print(news_soup.prettify())
```

```python
#Get the latest News Title and Paragraph Text and assign to variables
news_title = news_soup.find('div', class_='content_title').text.strip()
news_p = news_soup.find('div', class_='article_teaser_body').text.strip()

#Check result
print(news_title)
print(news_p)
```

### Mars Hemispheres

```python
#Read url
hem_url = 'https://marshemispheres.com/'
browser.visit(hem_url)
```

```python
#Create BeautifulSoup Object
hem_html = browser.html
hem_soup = bs(hem_html, 'html.parser')

# Retrieve all items that contain mars hemispheres information
items = hem_soup.find_all('div', class_='item')

# Create a list for hemisphere urls which will be appended later on
hemisphere_image_urls = []

# Loop through the items
for i in items: 
    # Store title
    title = i.find('h3').text
    
    # Store link that leads to full image for each hemisphere
    partial_img_url = i.find('a', class_='itemLink product-item')['href']
    
    # Visit the link that contains the full image website 
    browser.visit(hem_url + partial_img_url)
    
    #Create BeautifulSoup Object  
    partial_img_html = browser.html
    partial_soup = bs( partial_img_html, 'html.parser')
    
    # Retrieve full image source 
    img_url = hem_url + partial_soup.find('img', class_='wide-image')['src']
    
    # Append the retreived information into a list of dictionaries 
    hemisphere_image_urls.append({"title" : title, "img_url" : img_url})
    

# Display hemisphere_image_urls
hemisphere_image_urls
```
